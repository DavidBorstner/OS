```python
 #Seminarska naloga - Odlo훾itveni sistemi
from datetime import datetime as dt
import random
import collections
from operator import itemgetter
import math
```

```python
class UserItemData:
    
    def __init__(self, pot, from_date="1.1.1900", to_date="1.1.2100", min_ratings=0):
        self.pot = pot
        self.from_date = from_date
        self.to_date = to_date
        self.min_ratings = min_ratings
        self.SezVrstic = [] #Vse vrstice, v fajlu glede na datum
        self.PreberiRatinge() #Prebere vse vrstice za dolo훾ene datume
        
        self.StOcen = {} #Slovar stevila ocen za vsak film
        self.PrestejOceneFilmov()
        
        self.UstrezneOcene = [] #Vse ustrezne ocene glede na zahteve
        self.UstrezniFilmi() 
        
        self.PovprecjeVsehOcen = 0
        self.VsotaOcene = {}
        self.VsotaOcen()
        
        self.PovprecneOceneUporabnikov = {} #Slovar seznamov ocen za vsakega uporabnika posebej
        self.PovprecnaOcenaUporabnika()
        
        self.PovprecneOceneFilmov = {} #Slovar seznamov ocen za vsak film posebej
        self.PovprecnaOcenaFilma()
        
        
    def PreberiRatinge(self):
        self.SezVrstic = []
        data = open(self.pot)
        next(data)
        for line in data:
            vrsta = line.strip("\n").split("\t")
            datum = vrsta[3] + "." + vrsta[4] + "." + vrsta[5]
            vrsta = vrsta[:len(vrsta)-6]
            vrsta.append(datum)
            if dt.strptime(datum, "%d.%m.%Y") >= dt.strptime(self.from_date, "%d.%m.%Y") and  dt.strptime(datum, "%d.%m.%Y") < dt.strptime(self.to_date, "%d.%m.%Y"):
                self.SezVrstic.append(vrsta)
        data.close()
        
    def PrestejOceneFilmov(self):
        for i in self.SezVrstic:
            idU, idF, oc, dat = i
            if idF in self.StOcen:
                self.StOcen[idF] = self.StOcen[idF] + 1
            else:
                self.StOcen[idF] = 1
                
    def UstrezniFilmi(self):
        for i in self.SezVrstic:
            idU, idF, oc, dat = i
            if self.StOcen[idF] > self.min_ratings:
                self.UstrezneOcene.append(i)
                
    def nratings(self):
        return(len(self.UstrezneOcene))
    
    
    def VsotaOcen(self):
        data = open(self.pot)
        next(data)
        for line in data:
          vrsta = line.strip("\n").split("\t")
          if vrsta[1] in self.VsotaOcene:
                self.VsotaOcene[vrsta[1]] += float(vrsta[2])
                self.PovprecjeVsehOcen += float(vrsta[2])
          else:
                self.VsotaOcene[vrsta[1]] = float(vrsta[2])
        data.close()
        self.PovprecjeVsehOcen = self.PovprecjeVsehOcen / len(self.SezVrstic)
       
    def PovprecnaOcenaFilma(self):
        self.PovprecneOceneFilmov = {}

        for i in self.UstrezneOcene:
            idU,idF,oc,dat = i
            if idF in self.PovprecneOceneFilmov:
                self.PovprecneOceneFilmov[idF].append(float(oc))
            else:
                self.PovprecneOceneFilmov[idF] = [float(oc)]

    
    def PovprecnaOcenaUporabnika(self):
        self.PovprecneOceneUporabnikov = {}

        for i in self.UstrezneOcene:
            idU,idF,oc,dat = i
            if idU in self.PovprecneOceneUporabnikov:
                self.PovprecneOceneUporabnikov[idU].append(float(oc))
            else:
                self.PovprecneOceneUporabnikov[idU] = [float(oc)]

        
class Uporabnik: #Pomo탑ni razred za uporabnike
    
    def __init__(self,uid,uporabnik_id):
        self.uid = uid
        self.uporabnik_id = uporabnik_id
        self.OceneFilmov = {}
        self.PovprecnaOcena = 0
        
        self.RacunOcen()
        
    def RacunOcen(self):
        for i in self.uid.UstrezneOcene:
            idU,idF,oc,dat = i
            if int(idU) == int(self.uporabnik_id):
                self.PovprecnaOcena += float(oc)
                self.OceneFilmov[idF] = oc
        
        if len(self.OceneFilmov) <= 0:
            self.PovprecnaOcena = 0
        else:
            self.PovprecnaOcena = self.PovprecnaOcena/len(self.OceneFilmov)
    
class Film: #Pomo탑ni razred za filme
    
    def __init__(self,uid, fid):
        self.uid = uid
        self.fid = fid
        self.OceneFilmov = {}
        self.PovprecnaOcena = 0
        
        self.RacunOcen()
        
    def RacunOcen(self):
        for i in self.uid.UstrezneOcene:
            idU,idF,oc,dat = i
            if int(idF) == int(self.fid):
               self.PovprecnaOcena += float(oc)
               self.OceneFilmov[idU] = oc
                
        if len(self.OceneFilmov) <= 0:
            self.PovprecnaOcena = 0
        else:
            self.PovprecnaOcena = self.PovprecnaOcena/len(self.OceneFilmov)

class Filmi: 
    
    def __init__(self,pot):
        self.pot = pot
        self.SlovarFilm = {}
        self.SezFilm = []
        self.PreberiFilme()
        
    
    def PreberiFilme(self):   
        data = open(self.pot)
        next(data)
        for line in data:
            vrsta = line.strip("\n").split("\t")
            self.SezFilm.append(vrsta)
            if vrsta[0] not in self.SlovarFilm:
                self.SlovarFilm[int(vrsta[0])] = vrsta[1]
        data.close()
        
    def Naslov(self,id):
        return(self.SlovarFilm[id])
```

```python
## PREDIKTORJI

class RandomPredictor:
    
    def __init__(self, min,max):
        self.min = min
        self.max = max
        self.model = None
        self.napovedi = {}
    
    def fit(self,x):
        self.model = x
        
    def nauci(self):
        self.napovedi = {}
        for i in self.model.StOcen:
            self.napovedi[int(i)] = self.nakljucje()
    
    def predict(self,user_id):
        self.nauci()
        return(self.napovedi)
    
    def nakljucje(self):
        return (random.randint(self.min,self.max))
        
        
class AveragePredictor:
    
    def __init__(self,b):
        self.b = b
        self.model = None
        self.napovedi = {}
        
    def fit(self,x):
        self.model = x
        self.napovedi = {}
        for i in x.StOcen:
            self.napovedi[int(i)] =  self.povprecje(x.VsotaOcene[i], x.StOcen[i],x.PovprecjeVsehOcen)
            
    def povprecje(self,vs,n,g_avg):
        return (vs + self.b * g_avg) / (n + self.b)
        
    def predict(self, user_id):
        return(self.napovedi)

class Recomender:
    
    def __init__(self, p):
        self.prediktor = p # Prediktor
        self.napovedi = None
        self.model = None
        
    def fit(self, x):
        self.model = x
        pr = self.prediktor.fit(self.model) 
        
        
    #Seznam vseh k so ocenili filme v obeh obdobjih

    def evaluate(self, test_data, n):
        sezSkupnih = []

        for i in test_data.PovprecneOceneUporabnikov:
            if i in self.model.PovprecneOceneUporabnikov and i not in sezSkupnih:
                sezSkupnih.append(i)
        
        for i in self.model.PovprecneOceneUporabnikov:
            if i in test_data.PovprecneOceneUporabnikov and i not in sezSkupnih:
                sezSkupnih.append(i)
        
        SlvTo = {}
        
        for i in self.model.SezVrstic:
            idU, idF, oc, dat = i
            if idU in sezSkupnih:
                if idU not in SlvTo and idU in sezSkupnih:
                    SlvTo[idU] = {idF : oc} 
                else:
                    SlvTo[idU][idF] = oc
        
        
        SlvFrom = {}

        for i in test_data.SezVrstic:
            idU, idF, oc, dat = i
            if idU in sezSkupnih:
                if idU not in SlvFrom:
                    SlvFrom[idU] = {idF : oc} 
                else:
                    SlvFrom[idU][idF] = oc
        
        stevec = 0
        imenovalec = 0

        SlvNapovedi = {}
        self.prediktor.fit(self.model)

        for i in sezSkupnih:
            SlvNapovedi[i] = self.prediktor.predict(str(i))

        stevecMAE = 0
        stevecRMSE = 0
        imenovalec = 0

        TP = 0 #힋tevilo pravilnih napovedi
        FP = 0 #힋tevilo nepravilnih napovedi
        FN = 0 #힋tevilo filmov, ki jih nisem napovedal

    
        for i in SlvNapovedi:
            sezn = list(SlvNapovedi[i].values())    
            avg = 0
    
            stevecTP = 0
    
            if len(sezn) > 0:
                avg = sum(sezn)/len(sezn)
            for e in SlvNapovedi[i]:
                if e in SlvFrom[i]:
                    stevecMAE += abs(float(SlvFrom[i][e]) - float(SlvNapovedi[i][e]))
                    stevecRMSE += (float(SlvFrom[i][e]) - float(SlvNapovedi[i][e]))**2
                    imenovalec += 1
            
                if SlvNapovedi[i][e] >= avg and e in SlvFrom[i]:
                    TP += 1
                    stevecTP += 1
                    sezn.pop()
    
            FN += len(list(SlvFrom[i].values())) - stevecTP  # Vzamem 큄tevilo vseh pogledanih od dolo훾enega datuma, ter od큄tejem 큄tevilo filmov ki sem jih prav napovedal
            FP += len(sezn)
      
        
        MAE = stevecMAE/imenovalec
        RMSE = math.sqrt(stevecRMSE/imenovalec)
        natancnost = TP/(FP+TP)       
        priklic = TP/(TP+FN)
        fena = (2*priklic*natancnost/(priklic+natancnost))
        return([RMSE, MAE, natancnost, priklic, fena])
        
        
    def recommend(self, userID, n, seen):
        self.napovedi = self.prediktor.predict(str(userID))
        vrni = None
        if seen:
            sorted_x = sorted(self.napovedi.items(), key=lambda kv: kv[1], reverse=True)
            self.napovedi = collections.OrderedDict(sorted_x)
        
            top = list(self.napovedi.items())[:n]
        
            s = {}    
            for i in top:
                s[i[0]] = i[1]
        
            return(s)
        else:
            SlvSlv = {}
        
            for i in self.model.SezVrstic:
                idU, idF, oc, dat = i
                if idU not in SlvSlv:
                    SlvSlv[idU] = {idF : oc} 
                else:
                    SlvSlv[idU][idF] = oc
            
            for i in SlvSlv[str(userID)]:
                if i in self.napovedi:
                    del self.napovedi[i] 
            sorted_x = sorted(self.napovedi.items(), key=lambda kv: kv[1], reverse=True)
            self.napovedi = collections.OrderedDict(sorted_x)
              
            top = list(self.napovedi.items())[:n]
        
            s = {}    
            for i in top:
                s[i[0]] = i[1]
        
            return(s)
    
class ViewsPredictor:
    
    def __init__(self):
        self.model = None
        self.napovedi = {}
    
    def fit(self,x):
        self.model = x
    
    def predict(self,id):
        
        return(self.model.StOcen)
        
    
```

```python
#Skupinsko filtriranje

class ItemBasedPredictor: 
    
    def __init__(self, min_values=0, threshold=0):
        self.min_values = min_values
        self.threshold = threshold
        self.model = None
        self.pred = 0
        self.user_id = 0
        self.napovedi = None
        
    def fit(self, x):
        self.model = x
        
    def predict(self, usrID):   
        prvi = Ocenjevanje(self.model)
        self.napovedi = prvi.izracun(usrID)
        return(self.napovedi)
        
        
    def similarItems(self, id, n):
        movies = []

        for i in self.model.UstrezneOcene:
            if i[1] not in movies:
                movies.append(i[1])
                

        slv = {}

        for i in movies:
            if int(i) != id:
                slv[i] = self.similarityFilm(id, int(i))
                
        sorted_x = sorted(slv.items(), key=lambda kv: kv[1], reverse=True)
        napovedi = collections.OrderedDict(sorted_x)
        
        top = list(napovedi.items())[:20]
        
        s = {}    
        for i in top:
            s[i[0]] = i[1]
        return(s)

        
    def similarityFilm(self, p1, p2):
        flm1 = Film(self.model, p1)
        flm2 = Film(self.model, p2)
        
        ocenjevalci = []

        stevec = 0
        imenovalecLevo = 0
        imenovalecDesno = 0

        for i in flm1.OceneFilmov:
            if i in flm1.OceneFilmov and i in flm2.OceneFilmov and i not in ocenjevalci:
                ocenjevalci.append(i)
        

        for i in flm2.OceneFilmov:
            if i in flm1.OceneFilmov and i in flm2.OceneFilmov and i not in ocenjevalci:
                ocenjevalci.append(i)

        for i in ocenjevalci:
            avg = sum(self.model.PovprecneOceneUporabnikov[i])/len(self.model.PovprecneOceneUporabnikov[i]) #avg ocena ocenjevalca
            stevec += (float(flm1.OceneFilmov[i]) - avg) * (float(flm2.OceneFilmov[i]) - avg)
            imenovalecLevo += ((float(flm1.OceneFilmov[i]) - avg)**2)
            imenovalecDesno += ((float(flm2.OceneFilmov[i]) - avg)**2)
    
        zac =  (math.sqrt(imenovalecLevo) * math.sqrt(imenovalecDesno))
        if zac <= 0:
            zac = 1
        rezultat = (stevec/zac)
        
        if rezultat < self.threshold :
            return (0.0)
        else:
            return (rezultat)
            
    def similarityUser(self, u1, u2):
        usr1 = Uporabnik(self.model, str(u1))
        usr2 = Uporabnik(self.model, str(u2))
        
        ocenjevalci = []

        stevec = 0
        imenovalecLevo = 0
        imenovalecDesno = 0

        for i in usr1.OceneFilmov:
            if i in usr1.OceneFilmov and i in usr2.OceneFilmov and i not in ocenjevalci:
                ocenjevalci.append(i)
        

        for i in usr2.OceneFilmov:
            if i in usr1.OceneFilmov and i in usr2.OceneFilmov and i not in ocenjevalci:
                ocenjevalci.append(i)

        for i in ocenjevalci:
            avg = sum(self.model.PovprecneOceneFilmov[i])/len(self.model.PovprecneOceneFilmov[i]) #avg ocena ocenjevalca
            stevec += (float(usr1.OceneFilmov[i]) - avg) * (float(usr2.OceneFilmov[i]) - avg)
            imenovalecLevo += ((float(usr1.OceneFilmov[i]) - avg)**2)
            imenovalecDesno += ((float(usr2.OceneFilmov[i]) - avg)**2)
    
        zac =  (math.sqrt(imenovalecLevo) * math.sqrt(imenovalecDesno))
        if zac <= 0:
            zac = 1
        rezultat = (stevec/zac)
        
        if rezultat < self.threshold :
            return (0.0)
        else:
            return (rezultat)
        
        
# Class za ocenjevanje ocen

class Ocenjevanje:
    
    def __init__(self,uim):
        self.uim = uim
        self.SlvSlv = None
        self.SlovarPodobnosti = None
        self.Napovedi = None
        
    def podobnosti(self, id_ena): # Izra훾una podobnosti med uporabniki
        
        self.SlovarPodobnosti = {}

        for id_dva in self.SlvSlv:
                sezIstihFilmov = []
                for i in self.SlvSlv[id_ena]:
                    if i not in sezIstihFilmov and i in self.SlvSlv[id_dva]:
                        sezIstihFilmov.append(i)
        
                for i in self.SlvSlv[id_dva]:
                    if i not in sezIstihFilmov and i in self.SlvSlv[id_ena]:
                        sezIstihFilmov.append(i)
        
                stevec = 0
                imEna = 0
                imDva = 0
                imenovalec = 0

                for i in sezIstihFilmov:
                    ena_avg = sum(self.uim.PovprecneOceneUporabnikov[id_ena])/len(self.uim.PovprecneOceneUporabnikov[id_ena])
                    dva_avg = sum(self.uim.PovprecneOceneUporabnikov[id_dva])/len(self.uim.PovprecneOceneUporabnikov[id_dva])
        
                    ena_ocena = float(self.SlvSlv[id_ena][i])
                    dva_ocena = float(self.SlvSlv[id_dva][i])
                    stevec += ((ena_ocena - ena_avg) * (dva_ocena - dva_avg))
                    imEna += ((ena_ocena - ena_avg)**2)
                    imDva += ((dva_ocena - dva_avg)**2)
    
                imenovalec = math.sqrt(imEna) * math.sqrt(imDva)
                if id_dva not in self.SlovarPodobnosti and stevec != 0 and imenovalec != 0 and id_ena != id_dva:
                    povp = stevec/imenovalec
                    self.SlovarPodobnosti[id_dva] = povp
    
        
    def nastavi(self): # Naredi slovar slovarja ocen za vsakega uporabnika posebej
        self.SlvSlv = {}
        
        for i in self.uim.UstrezneOcene:
            idU, idF, oc, dat = i
            if idU not in self.SlvSlv:
                self.SlvSlv[idU] = {idF : oc} 
            else:
                self.SlvSlv[idU][idF] = oc
                
    def izracun(self, id_ena):  #Izra훾una ocene za uporabnike
        self.nastavi()
        self.podobnosti(id_ena)
        
        self.napovedi = {}
        for i in Sez_neogledanih:
            ena_avg = sum(self.uim.PovprecneOceneUporabnikov[id_ena])/len(self.uim.PovprecneOceneUporabnikov[id_ena])
            stevec = 0
            imenovalec = 0
            for id_dva in self.SlovarPodobnosti:
                if str(i) in self.SlvSlv[id_dva]:
                    dva_avg = sum(self.uim.PovprecneOceneUporabnikov[id_dva])/len(self.uim.PovprecneOceneUporabnikov[id_dva])
                    stevec += float(self.SlovarPodobnosti[id_dva]) * (float(self.SlvSlv[id_dva][i]) - dva_avg)
                    imenovalec += float(self.SlovarPodobnosti[id_dva])

            if stevec != 0 and imenovalec != 0:
                rez = ena_avg + (stevec/imenovalec)
                self.napovedi[i] = rez
                
        return(self.napovedi)        

```

```python
class SlopeOnePredictor:
    def __init__(self):
        self.model = None
        self.SlvSlv = None
        self.napoved = None
        
    def fit(self,x):
        self.model = x
        self.SlvSlv = {}
        
        for i in self.model.UstrezneOcene:
            idU, idF, oc, dat = i
            if idU not in self.SlvSlv:
                self.SlvSlv[idU] = {idF : oc} 
            else:
                self.SlvSlv[idU][idF] = oc
    
    def predict(self,user_ena):
        self.napoved = {}
        user_ena = str(user_ena)
        sez_neocenjenih_filmov = []

        for i in self.model.PovprecneOceneFilmov:
            if i not in self.SlvSlv[user_ena]:
                sez_neocenjenih_filmov.append(i)

        for film_ena in sez_neocenjenih_filmov:
            sez_ocenjevalcev_filma = [] #seznam uporabnikov, ki so ocenili dolo훾en film
            average = 0
            stevec = 0
    
            for i in self.SlvSlv:
                if film_ena in self.SlvSlv[i]:
                    sez_ocenjevalcev_filma.append(i)
            
            for i in sez_ocenjevalcev_filma:
                for e in self.SlvSlv[user_ena]: #ocene uporabnika ki ga 탑elimo napovedati
                    if e in self.SlvSlv[i]: #ocene ocenjvalca e je ocena, ki jo imamo 
                        average += float(self.SlvSlv[user_ena][e]) + (float(self.SlvSlv[i][film_ena]) - float(self.SlvSlv[i][e])) # ocena skupnega filma uporabnika user_ena + (ocena filma ki ga napovedujemo od uporabnika i - ocena skupnega filma)
                        stevec += 1
            
            self.napoved[film_ena] = (average/stevec)
        return(self.napoved)
        
```

```python
#Seznam vseh k so ocenili filme v obeh obdobjih

def evaluate(self, test_data, n):
    sezSkupnih = []

    for i in test_data.PovprecneOceneUporabnikov:
        if i in self.model.PovprecneOceneUporabnikov and i not in sezSkupnih:
            sezSkupnih.append(i)
        
    for i in self.model.PovprecneOceneUporabnikov:
        if i in test_data.PovprecneOceneUporabnikov and i not in sezSkupnih:
            sezSkupnih.append(i)
        
    SlvTo = {}
        
    for i in self.model.SezVrstic:
        idU, idF, oc, dat = i
        if idU in sezSkupnih:
            if idU not in SlvTo and idU in sezSkupnih:
                SlvTo[idU] = {idF : oc} 
            else:
                SlvTo[idU][idF] = oc
        
        
    SlvFrom = {}

    for i in test_data.SezVrstic:
        idU, idF, oc, dat = i
        if idU in sezSkupnih:
            if idU not in SlvFrom:
                SlvFrom[idU] = {idF : oc} 
            else:
                SlvFrom[idU][idF] = oc
        
    stevec = 0
    imenovalec = 0

    SlvNapovedi = {}
    self.prediktor.fit(self.model)

    for i in sezSkupnih:
        SlvNapovedi[i] = self.prediktor.predict(str(i))

    stevecMAE = 0
    stevecRMSE = 0
    imenovalec = 0

    TP = 0 #힋tevilo pravilnih napovedi
    FP = 0 #힋tevilo nepravilnih napovedi
    FN = 0 #힋tevilo filmov, ki jih nisem napovedal

    
    for i in SlvNapovedi:
        sezn = list(SlvNapovedi[i].values())    
        avg = 0
    
        stevecTP = 0
    
        if len(sezn) > 0:
            avg = sum(sezn)/len(sezn)
        for e in SlvNapovedi[i]:
            if e in SlvFrom[i]:
                stevecMAE += abs(float(SlvFrom[i][e]) - float(SlvNapovedi[i][e]))
                stevecRMSE += (float(SlvFrom[i][e]) - float(SlvNapovedi[i][e]))**2
                imenovalec += 1
            
            if SlvNapovedi[i][e] >= avg and e in SlvFrom[i]:
                TP += 1
                stevecTP += 1
                sezn.pop()
    
        FN += len(list(SlvFrom[i].values())) - stevecTP  # Vzamem 큄tevilo vseh pogledanih od dolo훾enega datuma, ter od큄tejem 큄tevilo filmov ki sem jih prav napovedal
        FP += len(sezn)
      
        
    MAE = stevecMAE/imenovalec
    RMSE = math.sqrt(stevecRMSE/imenovalec)
    natancnost = TP/(FP+TP)       
    priklic = TP/(TP+FN)
    fena = (2*priklic*natancnost/(priklic+natancnost))
    return([RMSE, MAE, natancnost, priklic, fena])
```
